# 첫줄에 단어의 갯수 랑 차원의 정보가 들어 있음 50000개로 추림

```python
with open('mapped_en_1.txt','rb') as f:
    with open('mapped_en.txt','wb') as ref:
        for i in range(50001):
            if i==0:
                f.readline()
                ref.write(b'50000 300\n')
                continue
                
            else:
                ref.write(f.readline())
    
```
